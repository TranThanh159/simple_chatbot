```python
import numpy as np

class my_system:
  def __init__(self):
    self.database_question = faiss.IndexFlatIP(384)
    self.database_answer = []

  def get_feature(self, question):
    embed = model.encode(question)
    embed = np.expand_dims(embed, axis=0)
    embed = np.asarray(embed, dtype=np.float32)

    #normalize to calculate Cosine similarity
    faiss.normalize_L2(embed)
    return embed

  def add(self, question, answer):
    embed = self.get_feature(question)
    #add to database
    self.database_question.add(embed)
    self.database_answer.append(answer)
    print("add sucessfully")

  def query(self, question):
    embed = self.get_feature(question)

    D, I = self.database_question.search(embed, 1)
    return D[0][0], self.database_answer[I[0][0]]

```

#Add question and answer into database

```python
system = my_system()

for i in range(len(data['intents'])):
  for question in data['intents'][i]['patterns']:
    print(question)
    print(data['intents'][i]['responses'][0])
    system.add(question, data['intents'][i]['responses'][0])
```

#Testcase
